#Heart Disease Predictions

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt


df = pd.read_csv("Heart_Disease_Prediction.csv")
df
```

```python
numList = ['Age',
        'Sex',
        'Chest pain type',
        'BP',
        'Cholesterol',
        'FBS over 120',
        'EKG results',
        'Max HR',
        'Exercise angina',
        'ST depression',
        'Slope of ST',
        'Number of vessels fluro',
        'Thallium']

for name in numList:
    df[name].plot.hist();
df.describe()
```

```python
df.plot.scatter(x="Age",y="Max HR")
```

```python

import plotly.express as px



#df = px.data.tips()
fig = px.scatter(df,
                 x="Age",
                 y="Max HR",
                 trendline="ols")
fig.show()


# Assuming your DataFrame is named 'df'
# If your column names are different (e.g., 'target' instead of 'Heart Disease',
# or 'thal' instead of 'Thalium'), adjust the names accordingly.

# Create a crosstab to count occurrences of Thalium values for each Heart Disease status
crosstab_df = pd.crosstab(df["Heart Disease"], df["Thallium"])

# Plot the crosstab as a bar chart
ax = crosstab_df.plot(kind="bar", figsize=(10, 6))

# Add some helpful labels and title
plt.title("Heart Disease Frequency by Thalium Result")
plt.xlabel("Heart Disease (0=No Disease, 1=Disease)")
plt.ylabel("Frequency")
plt.legend(title="Thallium Result", loc="upper right")
plt.xticks(rotation=0) # Keeps the x-axis labels horizontal for readability

# Show the plot
plt.show()




```
